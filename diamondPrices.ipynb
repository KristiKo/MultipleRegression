```python
%matplotlib inline
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('diamonds.csv')

df.head()
```

We considered a dataset of diamonds which includes prices, carat, depth, x (width). Price is plotted against carat. We grouped carat in intervals. As predicted, price increases when carat increases.

```python
df1 = df[['carat','price']]
intervals = np.arange(0.20, 1.5, 0.2)
groups = df1.groupby(pd.cut(df1['carat'], intervals)).mean()
groups['price'].plot.line()
```

```python
scale = StandardScaler()
X = df[['carat', 'depth', 'x']]
y = df['price']

X[['carat', 'depth', 'x']] = scale.fit_transform(X[['carat', 'depth', 'x']].as_matrix())
print (X)

est = sm.OLS(y, X).fit()

est.summary()
```

We select 3 features (carat, depth and width) and we use Standardscaler to create a matrix of values.

The OLs regression returns some values including coefficient, standard error etc. The coefficient indicates which is the features the mostly impact the price. As we expected, this feature is 'carat'

```python
y.groupby(df.carat).mean()
```

```python
scaled = scale.transform([[2.5, 63.3, 4.20]])
print(scaled)
predicted = est.predict(scaled[0])
print(predicted)
```

Finally we predicted the price of a diamond with 2.5 carat, 63.3 depth and 4.20 width (x) using the scale.transform which takes those 3 parameters.
